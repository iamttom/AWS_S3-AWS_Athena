## Data analysis

### Subtask:
Perform exploratory data analysis to understand the data, identify patterns, and gain insights.

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("Descriptive statistics:")
display(df_2022_1.describe())

categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal', 'target']
print("\nValue counts for categorical columns:")
for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    display(df_2022_1[col].value_counts())

print("\nCorrelation matrix:")
display(df_2022_1.corr(numeric_only=True))

numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
df_2022_1[numerical_cols].hist(figsize=(10, 8))
plt.suptitle('Histograms of Numerical Columns', y=1.02)
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, col in enumerate(categorical_cols):
    plt.subplot(3, 3, i + 1)
    sns.countplot(x=col, data=df_2022_1)
    plt.title(f'Count Plot for {col}')
plt.tight_layout()
plt.show()
```
